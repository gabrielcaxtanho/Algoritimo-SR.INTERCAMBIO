```python
#Puxando a imagem e aplicando algumas propriedades para facilitar a leitura do aquivo
img = cv2.imread('/content/imagem_2023-05-20_171246792.png')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

```python
#Pedinfo ao OCR para fazer a verificação da imagem extraindo os caracteres
config_tesseract = '--tessdata-dir tessdata'
texto = pytesseract.image_to_string(rgb, lang='por', config=config_tesseract)
print(texto)
```

```python
config_tesseract = '--tessdata-dir tessdata --psm 6'
texto = pytesseract.image_to_string(rgb, lang='por', config=config_tesseract)
texto_ingles = pytesseract.image_to_string(rgb, lang='eng', config=config_tesseract)

# Traduzir o texto do inglês para o português com a biblioteca googletrans
translator = Translator()
texto_traduzido = translator.translate(texto_ingles, dest='pt').text

# Imprimir o texto traduzido
print(texto_traduzido)
```
